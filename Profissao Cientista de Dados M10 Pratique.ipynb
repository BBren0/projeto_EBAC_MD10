```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head(5))
```

```python
#Gráfico de Dispersão
plt.figure(figsize=(10, 7))
sns.jointplot(x='Nota', y='Desconto', data=df, kind='scatter')
plt.title('Nota x Desconto')
plt.xlabel('Nota')
plt.ylabel('Desconto')
plt.show()
```

```python
# Gráfico de Calor
df_corr = df[['Qtd_Vendidos_Cod', 'Preço']].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
plt.title('Correlação entre Qtd_Vendidos_Cod e Preço')
plt.show()
```

```python
# Gráfico de Barras
sns.barplot(x='Qtd_Vendidos_Cod', y='Preço', data=df)
plt.title('Quantidade Vendida x Preço')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Preço')
plt.show()
```

```python
# Grafico de Pizza
plt.figure(figsize=(10,7))
# Filtrar valores maiores que 2%
genero_counts = df['Gênero'].value_counts()
genero_counts = genero_counts[genero_counts / genero_counts.sum() > 0.02]

# Plotar gráfico de pizza
plt.pie(genero_counts.values, labels=genero_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentual de vendas por gênero')
plt.show()
```

```python
# Gráfico de Densidade
plt.figure(figsize=(10,7))
sns.kdeplot(df['Preço'], fill=True, color = '#863e9c')
plt.title('Densidade de Preço')
plt.xlabel('Preço')
plt.show()
```
